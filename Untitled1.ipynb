```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('k_data.csv')
df.head()
```

```python
df.tail()
```

```python
df.shape
```

___preprocess the data___here we can see the 3 types of data

```python
df.dtypes
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
label_encoders = {}
for col in cat_cols:
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col].astype(str))
```

```python
print(df.head()) #0 mean ckd
```

```python
print(df.tail()) #2 mean notckd
```

```python
#df = df.fillna(df.mean())
df=df.dropna()
```

```python
print(df)
```

___________devide our dataset_______________

```python
X=df.drop(columns=['classification','id'])
```

```python
y=df['classification']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

_____________________train the model_______________________

```python
from sklearn.linear_model import LogisticRegression #algorithm is LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix



```

```python
from sklearn.svm import SVC

model = SVC()
```

```python
#model = LogisticRegression()# Adjust the number as needed
```

```python
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")

# Assuming y_pred is your predicted values from the model
y_pred = model.predict(X_test)

# Print classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Print confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")

# Assuming y_pred is your predicted values from the model
y_pred = model.predict(X_test)

# Print classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Print confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
